```python
import requests
import datetime
import pprint, codecs
import pandas as pd
from collections import defaultdict, Counter 
import pickle 
import re

token = open('C:/Users/USER/Documents/token.txt', 'r').read()
ids = open('C:/Users/USER/Documents/магистерская/практическая часть/ids.txt', 'r').read().split(',')
ids = ['-'+str(x) for x in ids]
results = {}

for gid in ids:
    try:
        r = requests.post('https://api.vk.com/method/execute.textGet?id='+
                          str(gid)+'&offset='+str(0)+'&v='+'5.73'+
                          '&access_token='+token).json()
        print('part 1 completed')
        text = []
        for i in range(len(r['response'])):
            text.extend(r['response'][i]['text'])

        print('part 2 completed')
        from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
        from nltk.corpus import stopwords

        stopWords = set(stopwords.words('russian'))
        wordsFiltered = []
        for i in range(len(text)):
            words = word_tokenize(text[i])
            for w in words:
                if w not in stopWords and len(w) > 4:
                    wordsFiltered.append(w)

        print('part 3 completed')
        result = Counter(wordsFiltered).most_common(50)
        screen_name = requests.get('https://api.vk.com/method/groups.getById',
                                    params={'group_id':str(gid[1:]),
                        'access_token':token, 'v':'5.73'}).json()['response'][-1]['name']
        screen_name =re.sub('[^\w_.)( -]', '', screen_name)
        results[screen_name] = result
        print('part 4 completed, nullifying')
    except Exception as e:
        pass
    
with open('words_pickle', 'wb') as fp:
    pickle.dump(results, fp)

print('Done')
```
